### Task 1 -- Gradient Descent with Armijo's rule

```python
opt = Optimizer("GD")
```

```python
%%time
pt = opt.optimize(x0=np.array([-1, 1]), expression="2*x1**2 + x1*x2 + 3*x2**2", show=True, method="armijo",
                  armijo_params={"alpha": 5,
                                 "epsilon": .5,
                                 "theta": .5})

opt.optimizer.iterations
```
