### Task 1 -- Gradient Descent with Armijo's rule

```python
opt = Optimizer("GD")
```

```python
%%time
pt = opt.optimize(x0=np.array([-1.0, 1.0]), function="sin(x+y) + (x - y)**2 - 1.5*x + 2.5*y + 1", show=True, method="armijo",
                  armijo_params={"alpha": 5,
                                 "epsilon": .5,
                                 "theta": .5})

opt.optimizer.iterations
```

### Task-6 Simplex method

```python
from scipy.optimize import linprog
```

```python
x0 = [1, 0, 1, 0]
```

```python
minimize = [-1, 3, 5, 1]

eq_vals = [[1, 4, 4, 1], 
           [1, 7, 8, 2]]
eq_rhs = [5, 9]
```

```python
linprog(c=minimize, A_eq=eq_vals, b_eq=eq_rhs, method="revised simplex", x0=x0, options={"disp": True})
```
