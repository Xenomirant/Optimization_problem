### Task-4 Conditional Gradient

### Task-6 Simplex method

```python
from scipy.optimize import linprog
```

```python
x0 = [1, 0, 1, 0]
```

```python
minimize = [-1, 3, 5, 1]

eq_vals = [[1, 4, 4, 1], 
           [1, 7, 8, 2]]
eq_rhs = [5, 9]
```

```python
linprog(c=minimize, A_eq=eq_vals, b_eq=eq_rhs, method="revised simplex", x0=x0, options={"disp": True})
```
